```python
# Importa biblioteca Pandas
import pandas as pd

# Função para exibir dados de uma tabela em formato de DataFrame
def exibir_tabela(nome_tabela):
  cursor.execute(f"SELECT * FROM {nome_tabela};")
  rows = cursor.fetchall()
  colunas = [desc[0] for desc in cursor.description]
  df = pd.DataFrame(rows, columns=colunas)
  print(f"\nDados da tabela {nome_tabela}:")
  display(df)

# Exibe tabelas criadas
cursor.execute("SHOW TABLES;")
tables = cursor.fetchall()
print("Tabelas no banco de dados 'Escola':")
for table in tables:
  print(table[0])
  exibir_tabela(table[0])
```

```python
# Consulta dados na tabela Disciplinas
cursor.execute("SELECT * FROM Disciplinas;")
for row in cursor.fetchall():
    print(row)
```

**Reasoning**:
MySQL Server has been installed, but it needs to be started before it can be configured. I will attempt to start the MySQL service using the `service mysql start` command.



```python
import subprocess

try:
    subprocess.run(['service', 'mysql', 'start'], check=True, capture_output=True)
    print("MySQL service started successfully.")
except subprocess.CalledProcessError as e:
    print(f"Error starting MySQL service: {e}")
    print(f"Stdout: {e.stdout.decode()}")
    print(f"Stderr: {e.stderr.decode()}")
except FileNotFoundError:
    print("Error: 'service' command not found. Ensure MySQL is correctly installed and its service command is accessible.")
```

**Reasoning**:
Now that MySQL is installed and running, I will configure the database by creating the 'Escola' database, setting up the 'adm' user, and granting the necessary privileges. This will ensure that the Python script can connect and interact with the database.



```python
import subprocess

try:
    # Create the database
    subprocess.run(["mysql", "-e", "CREATE DATABASE Escola;"], check=True, capture_output=True)
    print("Banco de dados 'Escola' criado com sucesso.")

    # Create the user and set password
    subprocess.run(["mysql", "-e", "CREATE USER 'adm'@'localhost' IDENTIFIED BY 'senha';"], check=True, capture_output=True)
    print("Usuário 'adm' criado com sucesso.")

    # Grant privileges
    subprocess.run(["mysql", "-e", "GRANT ALL PRIVILEGES ON Escola.* TO 'adm'@'localhost';"], check=True, capture_output=True)
    print("Privilégios concedidos ao usuário 'adm'.")

    # Flush privileges to apply changes
    subprocess.run(["mysql", "-e", "FLUSH PRIVILEGES;"], check=True, capture_output=True)
    print("Privilégios atualizados com sucesso.")

except subprocess.CalledProcessError as e:
    print(f"Error during MySQL configuration: {e}")
    print(f"Stdout: {e.stdout.decode()}")
    print(f"Stderr: {e.stderr.decode()}")
except FileNotFoundError:
    print("Error: 'mysql' command not found. Ensure MySQL client is installed and accessible.")
```

**Reasoning**:
Now that the MySQL server is installed, running, and configured with the 'Escola' database and 'adm' user, the next step is to establish a Python connection to the database using `mysql.connector`. This is crucial to verify the setup and proceed with creating the tables as required by the subtask.



```python
import mysql.connector
from mysql.connector import Error

try:
    # Close existing connection if it exists
    if 'conn' in locals() and conn.is_connected():
        conn.close()
        print("Conexão existente fechada.")

    # Conecta ao banco de dados Escola
    conn = mysql.connector.connect(
        host="localhost",
        user="adm",
        password="senha",
        database="Escola"
    )
    # Imprime mensagens para confirmar a conexão
    if conn.is_connected():
        print("Conexão realizada com sucesso.")

    cursor = conn.cursor()
    print("Cursor criado com sucesso.")

except Error as e:
    print(f"Erro ao conectar ao MySQL: {e}")
```

**Reasoning**:
With the database connection and cursor established, I will now create the 'Alunos', 'Professores', and 'Disciplinas' tables using `cursor.execute()` and then commit these changes to the database.



```python
try:
    # Cria tabelas no bd Escola
    cursor.execute("CREATE TABLE Alunos (id INT AUTO_INCREMENT PRIMARY KEY, nome VARCHAR(100) NOT NULL, data_nascimento DATE NOT NULL, endereco VARCHAR(255) NOT NULL, serie INT NOT NULL);")
    cursor.execute("CREATE TABLE Professores (id INT AUTO_INCREMENT PRIMARY KEY, nome VARCHAR(100) NOT NULL, especialidade VARCHAR(100) NOT NULL, data_contratacao DATE NOT NULL);")
    cursor.execute("CREATE TABLE Disciplinas (id INT AUTO_INCREMENT PRIMARY KEY, nome VARCHAR(100) NOT NULL, carga_horaria INT NOT NULL);")
    conn.commit()
    print("Tabelas 'Alunos', 'Professores' e 'Disciplinas' criadas com sucesso.")
except Error as e:
    print(f"Erro ao criar tabelas: {e}")
    if conn and conn.is_connected():
        conn.rollback() # Rollback on error
```
